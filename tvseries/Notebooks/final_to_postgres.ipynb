```python
import requests
import json
from pprint import pprint
import pandas as pd
import numpy as np
```

```python
url='http://api.tvmaze.com/shows?page='
```

```python
page_list=np.arange(165)
page_list
```

```python
id_=[]
name=[]
url_=[]
language=[]
genre=[]
premiered=[]
rating=[]
country=[]
image=[]
summary=[]
updated=[]
error_list=[]

for n in page_list:
    response = requests.get(url+str(page_list[n])).json()
    print("Performing page "+str(n))
    for i in np.arange(len(response)):
        id_.append(response[i]["id"])
        name.append(response[i]["name"])
        url_.append(response[i]["url"])
        language.append(response[i]["language"])
        try:
            genre.append(response[i]["genres"][0])
        except:
            genre.append("Nan")
        premiered.append(response[i]["premiered"])
        rating.append(response[i]["rating"]["average"])
        try:
            country.append(response[i]["network"]["country"]["name"])
        except:
            country.append("NaN")
        try:
            image.append(response[i]["image"]["original"])
        except:
            image.append("NaN")
        summary.append(response[i]["summary"])
        updated.append(response[i]["updated"])
    
```

```python
print(len(genre))
```

```python
print(len(updated))
```

```python
# create dataframe
shows_df = pd.DataFrame({
    'id': id_,
    'name': name,
    'url': url_,
    'language':language,
    'genre':genre,
    'premiered':premiered,
    'rating':rating,
    'country':country,
    'image':image,
    'summary':summary,
    'updated':updated
    })

shows_df
```

```python
shows_df.to_csv('shows_df.csv')
```
